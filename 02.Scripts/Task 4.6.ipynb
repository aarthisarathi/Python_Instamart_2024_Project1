### 3.Determine a suitable way to combine the orders_products_combined dataframe with your products data set

```python
#As this df_ords_prods_combined already has _merge column so dropping that

df_ords_prods_combined = df_ords_prods_combined.drop(['_merge'], axis=1)
#check output
df_ords_prods_combined.head()
```

```python
df_ords_prods_combined['Unnamed: 0']
```

## I have optimized data to avoid memory error

```python
df_ords_prods_combined['order_id']=df_ords_prods_combined['order_id'].astype('int32')
df_ords_prods_combined['product_id']=df_ords_prods_combined['product_id'].astype('int32')
df_ords_prods_combined['add_to_cart_order']=df_ords_prods_combined['add_to_cart_order'].astype('int32')
df_ords_prods_combined['reordered']=df_ords_prods_combined['reordered'].astype('int32')
df_ords_prods_combined['order_number']=df_ords_prods_combined['order_number'].astype('int32')
df_ords_prods_combined['order_dow']=df_ords_prods_combined['order_dow'].astype('int32')
df_ords_prods_combined['order_dow']=df_ords_prods_combined['order_dow'].astype('int32')
df_ords_prods_combined['order_hour_of_day']=df_ords_prods_combined['order_hour_of_day'].astype('int32')
df_ords_prods_combined['days_since_prior_order']=df_ords_prods_combined['days_since_prior_order'].astype('float32')
df_ords_prods_combined['user_id']=df_ords_prods_combined['user_id'].astype('int32')
```

```python
df_ords_prods_combined.info()
```

```python
df_ords_prods_combined.head()
```

```python
df_ords_prods_combined.drop(['Unnamed: 0'], axis=1)
```

#### Merging df_ords_prods_combined with df_prods

```python
df_ords_prods_merge = df_prods.merge(df_ords_prods_combined , on = 'product_id' , indicator = True)
#check output
df_ords_prods_merge
```

### 4.Checking the counts of the result using the merge flag

```python
#Checking counts in merged dataframe
df_ords_prods_merge['_merge'].value_counts()
```

#### The resulting dataframe (after the merge) has 32,404,859 rows, and each of those rows have information found in both input data sets, as we used an inner join for the purposes of this project.

### 5.Export the newly created dataframe as ords_prods_merge in a pkl format

```python
df_ords_prods_merge.rename(columns={'order_dow': 'order_date_of_the_week'}, inplace=True)
```

```python
df_ords_prods_merge = df_ords_prods_merge.drop(['Unnamed: 0'], axis=1)
```

```python
df_ords_prods_merge = df_ords_prods_merge.drop(['eval_set'], axis=1)
```

df_ords_prods_merge.head()
