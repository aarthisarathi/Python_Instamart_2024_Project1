### Step 4

```python
# Creating new column "max_order"
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking for new column
df_ords_prods_merge.head(20)
```

```python
# Assign loyalty flag

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'

df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in dataframe
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number','max_order']].head(20)
```

### Step 5

Changing the outliers priced 99999 to np.nan

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# the min max and mean of each loyalty flag
df_ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean','min','max']})
```

New customers have the least price of products in both the average and max. Loyal customer have the highest price of products in both the average and max.

### Step 6

```python
#creating average_spending column to capture average of price across all orders

df_ords_prods_merge['average_spending'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
df_ords_prods_merge
```

```python
#setting flag for spending

df_ords_prods_merge.loc[df_ords_prods_merge['average_spending'] < 10, 'spending_flag'] = 'Low spender'

df_ords_prods_merge.loc[df_ords_prods_merge['average_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check spending flags
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Step 7

```python
#creating customer_frequency column to capture average of price across all orders

df_ords_prods_merge['customer_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
df_ords_prods_merge
```

```python
#setting flag for frequency

df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'

df_ords_prods_merge.loc[(df_ords_prods_merge['customer_frequency'] > 10) & (df_ords_prods_merge['customer_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'

df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
